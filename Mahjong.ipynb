```python
# 斷么九判斷
def checktanyao(x):
    global tanyao, havehan, player1_pon, player1_kan
    GG = 0
    for i in player1_pon:
        a = [0,1,2,10,11,19,20,21,22,23,24,25,26]
        b = int(i[1]) // 4
        if b in a:
            GG = 1
            break
    for i in player1_kan:
        a = [0,1,2,10,11,19,20,21,22,23,24,25,26]
        b = i[1] // 4
        if b in a:
            GG = 1
            break
    for i in player1_ankan:
        a = [0,1,2,10,11,19,20,21,22,23,24,25,26]
        b = i[1] // 4
        if b in a:
            GG = 1
            break
    for i in x:
        a = [0,1,2,10,11,19,20,21,22,23,24,25,26]
        b = i[1] // 4
        if b in a:
            GG = 1
            break
    if GG == 0:
        tanyao = 1
        havehan = True
```

```python
# 找所有可能雀頭
def findpair(x):
    tmp = []
    pair = []
    setx = set(x)
    for a in setx:
        for b in range(len(x)):
            if x[b] == a:
                tmp.append(a)
        if len(tmp) >= 2:
            pair.append(tmp[0])
            tmp.clear()
        else:
            tmp.clear()
    return pair
```

```python
# 變好看一點
def cleanhand(x):
    global hand
    for i in x:
        hand.append(i[0])
    return hand
```

```python
# 測最小牌能否找到順子
def checkflush(x):
    if x:
        if x[0] + 1 in x:
            tmp1 = x[0] + 1
            if x[0] + 2 in x:
                tmp2 = x[0] + 2
                x.remove(x[0])
                x.remove(tmp1)
                x.remove(tmp2)
    return x
```

```python
# 測最小牌能否找到暗刻
def checktriple(x):
    global tri
    tmp = []
    for a in range(len(x)):
        if x[a] == x[0]:
            tmp.append(x[0])
    if len(tmp) >= 3:
        x.remove(tmp[0])
        x.remove(tmp[0])
        x.remove(tmp[0])
        tri = 1
    return x
```

```python
# 測條子跟筒子是否會消光光
def roncheck(x):
    tri = 0
    win = 0
    count = 0
    while x:
        checktriple(x)
        if not x:
            win = 1
            return x
        if tri == 1:
            tri = 0
            continue
        checkflush(x)
        count += 1
        if not x:
            win = 1
            return x
        if count > 5:
            break
    if win == 0:
        return x
```

```python
# 測是不是全部都暗刻，萬子跟字牌專用
def checktriple2(x):
    tmp = []
    for a in x:
        for b in range(len(x)):
            if a == x[b]:
                tmp.append(a)
        if len(tmp) == 3:
            x.remove(tmp[0])
            x.remove(tmp[0])
            x.remove(tmp[0])
        else:
            tmp.clear()
    return x
```

```python
# 胡牌檢查
import copy
def checkron(x):
    global ron
    GG = 0
    check = []
    checkm = []
    checkp = []
    checks = []
    checky = []
    for i in x:
        check.append(int(i[1] // 4))
    for i in check:
        if i == 0:
            checkm.append(1)
        elif i == 1:
            checkm.append(9)
        elif 2 <= i <= 10:
            checkp.append(int(i - 1))
        elif 11 <= i <= 19:
            checks.append(int(i - 10))
        else:
            checky.append(int(i - 19))
    mpair = findpair(checkm)
    ppair = findpair(checkp)
    spair = findpair(checks)
    ypair = findpair(checky)
    for x in mpair:
        mtmp = copy.deepcopy(checkm)
        ptmp = copy.deepcopy(checkp)
        stmp = copy.deepcopy(checks)
        ytmp = copy.deepcopy(checky)
        mtmp.remove(x)
        mtmp.remove(x)
        checktriple2(mtmp)
        if mtmp:
            continue
        roncheck(ptmp)
        if ptmp:
            continue
        roncheck(stmp)
        if stmp:
            continue
        checktriple2(ytmp)
        if ytmp:
            continue
        else:
            ron = 1
            return ron
    for x in ppair:
        mtmp2 = copy.deepcopy(checkm)
        ptmp2 = copy.deepcopy(checkp)
        stmp2 = copy.deepcopy(checks)
        ytmp2 = copy.deepcopy(checky)
        ptmp2.remove(x)
        ptmp2.remove(x)
        checktriple2(mtmp2)
        if mtmp2:
            continue
        roncheck(ptmp2)
        if ptmp2:
            continue
        roncheck(stmp2)
        if stmp2:
            continue
        checktriple2(ytmp2)
        if ytmp2:
            continue
        else:
            ron = 1
            return ron
    for x in spair:
        mtmp3 = copy.deepcopy(checkm)
        ptmp3 = copy.deepcopy(checkp)
        stmp3 = copy.deepcopy(checks)
        ytmp3 = copy.deepcopy(checky)
        stmp3.remove(x)
        stmp3.remove(x)
        checktriple2(mtmp3)
        if mtmp3:
            continue
        roncheck(ptmp3)
        if ptmp3:
            continue
        roncheck(stmp3)
        if stmp3:
            continue
        checktriple2(ytmp3)
        if ytmp3:
            continue
        else:
            ron = 1
            return ron
    for x in ypair:
        mtmp4 = copy.deepcopy(checkm)
        ptmp4 = copy.deepcopy(checkp)
        stmp4 = copy.deepcopy(checks)
        ytmp4 = copy.deepcopy(checky)
        ytmp4.remove(x)
        ytmp4.remove(x)
        checktriple2(mtmp4)
        if mtmp4:
            continue
        roncheck(ptmp4)
        if ptmp4:
            continue
        roncheck(stmp4)
        if stmp4:
            continue
        checktriple2(ytmp4)
        if ytmp4:
            continue
        else:
            ron = 1
            return ron
    return ron
```
