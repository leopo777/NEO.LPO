```python
# 找所有可能雀頭
def findpair(x):
    tmp = []
    pair = []
    setx = set(x)
    for a in setx:
        for b in range(len(x)):
            if x[b] == a:
                tmp.append(a)
        if len(tmp) >= 2:
            pair.append(tmp[0])
            tmp.clear()
        else:
            tmp.clear()
    return pair
```

```python
# 測最小牌能否找到順子
def checkflush(x):
    if x:
        if x[0] + 1 in x:
            tmp1 = x[0] + 1
            if x[0] + 2 in x:
                tmp2 = x[0] + 2
                x.remove(x[0])
                x.remove(tmp1)
                x.remove(tmp2)
    return x
```

```python
# 測最小牌能否找到暗刻
def checktriple(x):
    global tri
    tmp = []
    for a in range(len(x)):
        if x[a] == x[0]:
            tmp.append(x[0])
    if len(tmp) >= 3:
        x.remove(tmp[0])
        x.remove(tmp[0])
        x.remove(tmp[0])
        tri = 1
    return x
```

```python
# 測條子跟筒子是否會消光光
def roncheck(x):
    tri = 0
    win = 0
    count = 0
    while x:
        checktriple(x)
        if not x:
            win = 1
            return x
        if tri == 1:
            tri = 0
            continue
        checkflush(x)
        count += 1
        if not x:
            win = 1
            return x
        if count > 5:
            break
    if win == 0:
        return x
```
